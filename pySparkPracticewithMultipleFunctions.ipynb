```
#Create DataFrame and Rows manually

from pyspark.sql import Row

row1=Row('Viswanath',32,'Male')
row2=Row('Prasanna', 32,'Female')

createSchema='Name string, Age integer, Gender string'

df2=spark.createDataFrame([row1, row2], createSchema)
df2.show()
```

```
df2.collect()
```

```
#Select function -- I will show the selected row's data

from pyspark.sql.functions import expr, col
df3=df2.select("Name", "Age")
df3.show()
```

```
#Literals Function - lit() = used to add a new constant column to DataFrame by assigning a literal or constant value

from pyspark.sql.functions import lit

df3.select(expr("*"), lit(1).alias("RollNo")).show(2)
df3.select(expr("*"), lit(2).alias("RollNo")).show(2)
df3.select(expr("*"), lit(3).alias("RollNo")).show(2)
```
